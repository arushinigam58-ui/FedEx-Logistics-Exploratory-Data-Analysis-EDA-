```python
# 1. Clean Freight Cost: Convert text to NaN, then fill with median or 0
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')
median_freight = df['Freight Cost (USD)'].median()
df['Freight Cost (USD)'] = df['Freight Cost (USD)'].fillna(median_freight, inplace=True)
```

```python
# 2. Clean Weight: Convert text to NaN, then fill with median
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'], errors='coerce')
median_weight = df['Weight (Kilograms)'].median()
df['Weight (Kilograms)'] = df['Weight (Kilograms)'].fillna(median_weight, inplace=True)
```

```python
# 3. Create a Delivery Delay feature (Crucial for your business questions!)
# Convert dates to datetime formats
df['PO Sent to Vendor Date'] = pd.to_datetime(df['PO Sent to Vendor Date'], errors='coerce')
df['Scheduled Delivery Date'] = pd.to_datetime(df['Scheduled Delivery Date'], errors='coerce')
df['Delivered to Client Date'] = pd.to_datetime(df['Delivered to Client Date'], errors='coerce')
```

```python
# calculate 'Days_To_Schedule'
df['Days_To_Schedule'] = (df['Scheduled Delivery Date'] - df['PO Sent to Vendor Date']).dt.days
median_days = df['Days_To_Schedule'].median()
df['Days_To_Schedule']= df['Days_To_Schedule'].fillna(median_days)
```

```python
#clear nat from po sent to vendor date 
df['PO Sent to Vendor Date'] = df['PO Sent to Vendor Date'].fillna(df['Scheduled Delivery Date'] - pd.to_timedelta(median_days, unit='D'))
```

```python
# Calculate days gap and if a shipment was delayed
df['Days_To_Schedule'] = (df['Scheduled Delivery Date'] - df['PO Sent to Vendor Date']).dt.days
df['Is_Delayed'] = df['Delivered to Client Date'] > df['Scheduled Delivery Date']
```

```python
print('data Wragling Complete Sucessfuly!')
print(df[['Days_To_Schedule' , 'Is_Delayed','Weight (Kilograms)', 'Freight Cost (USD)','PO Sent to Vendor Date']].isna().sum())
```

```python
# Fill shipment mode with most frequent value 
df['Shipment Mode'] = df['Shipment Mode'].fillna(df['Shipment Mode'].mode()[0])
```

```python
# Fill Dosage with most frequent value 
df['Dosage'] = df['Dosage'].fillna(df['Dosage'].mode()[0])
```

```python
# Fill Line Item Insurance (USD) with most frequent value 
df['Line Item Insurance (USD)'] = df['Line Item Insurance (USD)'].fillna(df['Line Item Insurance (USD)'].mode()[0])
```

```python
# Verify no missing values remain
print(df.isnull().sum())
```

# Milestone Answers 


Milestone 1: Do shipments from certain countries experience more delays compared to others?

Answer: Yes. Based on cross-tabulating countries against delivery delays, landlocked destinations and regions with developing logistics infrastructure (such as Côte d'Ivoire and Nigeria) exhibit a higher frequency of delayed arrivals. Conversely, countries with highly centralized supply channels like South Africa show significantly better on-time performance metrics.


Milestone 2: Does the shipment mode impact the frequency of on-time deliveries?

Answer: Heavily. Air Transport maintains the most consistent track record for meeting scheduled delivery dates due to dedicated commercial routes. Trucking exhibits moderate delay frequencies due to unpredictable regional infrastructure and border customs checks. Ocean Freight, while cost-effective for bulk shipments, suffers from the longest delivery lead times and the highest variance in schedules.


Milestone 3: Is there a difference in delivery performance based on the time between the PO Sent to Vendor Date and the Scheduled Delivery Date?

Answer: Yes, a clear threshold exists. Orders with narrow scheduling windows (short lead times between the PO being sent and the expected delivery date) show a massive spike in delays. Vendors require a baseline operational window; compressing this timeframe dramatically decreases the probability of an on-time arrival.


Milestone 4: Does the type of INCO term used impact vendor delivery performance?

Answer: Yes. Shipments routed under EXW (Ex Works) place the logistics burden entirely on the buyer immediately upon leaving the factory gates, leading to unpredictable transit delays. Terms like RDC (Regional Distribution Centers) reflect highly optimized, pre-negotiated lanes that deliver superior, predictable timing performance.


Milestone 5: Are shipments with higher weights more likely to incur higher Insurance costs?

Answer: No. The data reveals a very weak correlation between shipment weight and insurance fees. Insurance costs scale linearly with Line Item Value, not weight. A small package of high-value medications costs substantially more to insure than a heavy pallet of low-cost consumables.

 ## 5. Solution to Business Objective;

To achieve the core business objectives of optimizing costs and minimizing supply chain bottlenecks, FedEx Logistics should implement three targeted strategies:

Automated Transport Mode Switching: Integrate a rule-based engine in the order placement system. When cargo weights exceed a specific threshold and timelines allow, automatically shift bookings from Air to Ocean or Trucking to cut down on premium freight expenditure.

Data-Entry Standardization: Enforce strict numeric-only front-end constraints on the Freight Cost and Weight input fields in the logistics portal. This will instantly eliminate corrupted text variables (like "Weight Captured Separately") and prevent operational blindspots.

Strategic Buffer Allocation: Mandate a minimum lead-time window between the "PO Sent Date" and "Scheduled Delivery Date" for high-risk lanes (e.g., specific African destinations under EXW terms) to systematically eliminate rush-order bottlenecks.
